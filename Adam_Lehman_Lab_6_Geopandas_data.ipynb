# Geopandas

Let's start off where we ended last week. Load the city of Denver neighborhood data and make a map running the code below.

An geospatial extension of pandas is geopandas, which you can learn more about here: https://geopandas.org/en/stable/

Geopandas stores data in rows and columns like pandas does, but it also includes geometry data that can be mapped. We'll dive more into Geopandas in the future. But for now, let's try a short example.

<mark background-color="lightyellow">Question 1. Run the code below. What does it do?</mark>



```python
import geopandas as gpd

denver_data = gpd.read_file("https://www.denvergov.org/media/gis/DataCatalog/census_neighborhood_demographics_2010/shape/census_neighborhood_demographics_2010.zip")
denver_data.plot("POPULATION",legend=True)


# Question 1. This code retrieves data from denvergov.org and displys a demographic map by means of geopandas.
```

<mark background-color="lightyellow">Question 9. Make a centroid for the auraria neighborhood using the Colorado Central State Plane feet CRS, and print the dataframe.</mark>

<mark background-color="lightyellow">CHALLENGE PROBLEM. Question 10. Make a centroid of the Washington Park neighborhood in feet, and measure the distance between the Auraria and the Washington Park centroids.</mark>

```python
# Question 10. The distance between Auraria and Washington Park is 20,082.82 feet.

# Create a dataframe of Washington Park
washpark = denver_data.loc[['Washington Park']]

# Convert to feet
washpark_ft = washpark.to_crs(2232)

# Create centroid
washpark_centroid = washpark_ft.centroid

# Calculate distance
distance = washpark_centroid[0].distance(auraria_centroid[0])
distance
```
